```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\Chris\\Downloads\\plane_crashes_data.csv")
```

```python
# matplotlib histogram
crashes5060 = df.loc[(df['year']>=1950)&(df['year']<1960), 'fatalities']
crashes6070 = df.loc[(df['year']>=1960)&(df['year']<1970), 'fatalities']
crashes7080 = df.loc[(df['year']>=1970)&(df['year']<1980), 'fatalities']
crashes8090 = df.loc[(df['year']>=1980)&(df['year']<1990), 'fatalities']
crashes9000 = df.loc[(df['year']>=1990)&(df['year']<2000), 'fatalities']

plt.hist(crashes5060, color='red', alpha=0.5)
plt.hist(crashes6070, color='yellow', alpha=0.5)
plt.hist(crashes7080, color='blue', alpha=0.5)
plt.hist(crashes8090, color='green', alpha=0.5)
plt.hist(crashes9000, color='purple', alpha=0.5)
plt.show()
```

```python
newdata = df.groupby(['month', 'year']).sum().reset_index()

# One Variable - Plane crash fatalities
# 1. Histogram
sns.distplot(newdata.fatalities, hist=True)
plt.show()

# 2. Line Plot
plt.plot(newdata.fatalities)
plt.show()

# 3. Box Plot
sns.boxplot(x=None, y=newdata.fatalities, color='purple')
plt.show()

# Seoborn Bar Plot
sns.barplot(x=None, y=newdata.fatalities, color='r')
plt.show()
```

```python
# Two variables - Plane crash fatalities & Year
# 1. Seaborn Box Plot
sns.boxplot(x=newdata.year, y=newdata.fatalities)
plt.show()

# 2. Seanborn Line Plot
sns.lineplot(x=newdata.year, y=newdata.fatalities, data=newdata)
plt.show()

# 3. Strip Plot
sns.stripplot(x=newdata.year, y=newdata.fatalities, data=newdata)
plt.show()
```

```python
# Continuous Variables

# 1. Seanborn Scatter Plot
sns.scatterplot(x=newdata.aboard, y=newdata.fatalities, data=newdata)
plt.show()

#2 Joint Kernal Density
sns.jointplot(x=newdata.aboard, y=newdata.fatalities, data=newdata, kind="kde")
plt.show()

#3 Hex plot
sns.jointplot(x=newdata.aboard, y=newdata.fatalities, kind="hex", color="green")
plt.show()

#4 Seaborn heatmap
pivotchart = newdata.pivot(index='month', columns='year', values='fatalities')
sns.heatmap(pivotchart, linewidths=.5)
plt.show()

#5 Vertical Bar plots by year
sns.barplot(y=newdata.fatalities, x=newdata.year, data=newdata,
            label="Fatalities", color="b")
sns.barplot(y=newdata.aboard, x=newdata.year, data=newdata,
            label="Aboard", color="b")
plt.show()
```
